```python
import sys
sys.path.append('../src')
from popgen import *
%matplotlib inline
from ipywidgets import interact
```

## Simple example - Biallelic

```python
def drift_pop(num_msats, pop_size):
    m = SinglePop(gens=100)
    m.num_msats = num_msats
    m.pop_size = pop_size
    BasicView(m, [ExpHe()], ['mean'], min_y=[0], max_y=[1])
    m.run()
    
interact(drift_pop, num_msats=(2, 20, 1), pop_size=(30, 530, 50))
```

```python
m = SinglePop(gens=100)
m.num_snps = 11
m.pop_size = 50
BasicView(m, [ExpHe()], ['mean'])
m.run()
```

```python
m = SinglePop(gens=100)
m.num_snps = 5
m.snp_freq = [0.01, 0.05, 0.1, 0.5, 0.9]
m.pop_size = 50
BasicView(m, [ExpHe()], ['mean'])
m.run()
```

## Simple example - Microsatellites

```python
m = SinglePop(gens=100)
m.num_msats = 10
m.pop_size = 50
BasicView(m, [ExpHe(), NumAlleles()], ['mean'])
m.run()
```

```python
m = SinglePop(gens=100)
m.num_msats = 10
m.pop_size = 50
m.num_msats = [2, 5, 10, 20]
BasicView(m, [ExpHe(), NumAlleles()], ['mean'])
m.run()
```
